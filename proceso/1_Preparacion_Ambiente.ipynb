```sql
%python
dbutils.widgets.removeAll()
```

```sql
create widget text storageName default "adlssmartdatajuanfinal";
create widget text storageLocation default "abfss://lhcldata@adlssmartdatajuanfinal.dfs.core.windows.net";
```

```sql
CREATE CATALOG IF NOT EXISTS catalog_dev;
```

```sql
CREATE SCHEMA IF NOT EXISTS catalog_dev.bronze;
CREATE SCHEMA IF NOT EXISTS catalog_dev.silver;
CREATE SCHEMA IF NOT EXISTS catalog_dev.golden;
```

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS `exlt-raw`
URL 'abfss://raw@${storageName}.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL credential)
```

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS `exlt-bronze`
URL 'abfss://bronze@${storageName}.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL credential)
COMMENT 'Ubicación externa para las tablas bronze del Data Lake';
```

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS `exlt-silver`
URL 'abfss://silver@${storageName}.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL credential)
COMMENT 'Ubicación externa para las tablas silver del Data Lake';
```

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS `exlt-golden`
URL 'abfss://golden@${storageName}.dfs.core.windows.net/'
WITH (STORAGE CREDENTIAL credential)
COMMENT 'Ubicación externa para las tablas golden del Data Lake';
```

### Creacion tablas Bronze

```sql
CREATE OR REPLACE TABLE catalog_dev.bronze.Bitacora_incidentes (
  Cod_incidente STRING,
  Zona STRING,
  Estado STRING,
  Hora_Asignacion TIMESTAMP,
  Tipo_Vehiculo STRING,
  Numero_Tecnicos INT,
  Equipo_Afectado STRING,
  Tipo_Falla STRING,
  Prioridad INT,
  Lider_Tecnico STRING,
  Fecha_Ejecucion TIMESTAMP
)
USING DELTA
LOCATION '${storageLocation}/bronze/Bitacora_incidentes';


CREATE OR REPLACE TABLE catalog_dev.bronze.Maestro_horas (
  Zona STRING,
  Criticidad STRING,
  Tiempo_Espera_Horas INT,
  Fecha_Ejecucion TIMESTAMP
)
USING DELTA
LOCATION '${storageLocation}/bronze/Maestro_horas';
```

### Creacion tabla Silver

```sql
CREATE OR REPLACE TABLE catalog_dev.silver.Bitacora_clasificacion (
  Cod_incidente STRING,
  Zona STRING,
  Estado STRING,
  Hora_Asignacion TIMESTAMP,
  Tipo_Vehiculo STRING,
  Numero_Tecnicos INT,
  Equipo_Afectado STRING,
  Tipo_Falla STRING,
  Prioridad INT,
  Lider_Tecnico STRING,
  Criticidad STRING,
  Tiempo_Espera_Horas INT,
  Horas_Transcurridas INT,
  Clasificacion STRING,
  Fecha_Ejecucion TIMESTAMP
)
USING DELTA
LOCATION '${storageLocation}/silver/Bitacora_clasificacion';
```

### Creacion tablas Gold

```sql
--DROP TABLE IF EXISTS catalog_dev.golden.clasificacion_vehiculo;
CREATE OR REPLACE TABLE catalog_dev.golden.clasificacion_vehiculo (
  Tipo_Vehiculo STRING,
  Clasificacion STRING,
  Cantidad LONG,
  Fecha_Ejecucion TIMESTAMP
)
USING DELTA
LOCATION '${storageLocation}/golden/clasificacion_vehiculo';


--DROP TABLE IF EXISTS catalog_dev.golden.ranking_lideres;
CREATE OR REPLACE TABLE catalog_dev.golden.ranking_lideres (
  Lider_Tecnico STRING,
  Total_Incidentes LONG,
  Fecha_Ejecucion TIMESTAMP
)
USING DELTA
LOCATION '${storageLocation}/golden/ranking_lideres';


--DROP TABLE IF EXISTS catalog_dev.golden.tendencia_diaria;
CREATE OR REPLACE TABLE catalog_dev.golden.tendencia_diaria (
  Incidentes_por_dia LONG,
  Fecha_Ejecucion TIMESTAMP
)
USING DELTA
LOCATION '${storageLocation}/golden/tendencia_diaria';

--DROP TABLE IF EXISTS catalog_dev.golden.zona_estado;
CREATE OR REPLACE TABLE catalog_dev.golden.zona_estado (
  Zona STRING,
  Total_Incidentes_Zona LONG,
  Fecha_Ejecucion TIMESTAMP
)
USING DELTA
LOCATION '${storageLocation}/golden/zona_estado';
```
